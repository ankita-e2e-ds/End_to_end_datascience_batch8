```python
# 2. In wine quality dataset, plot a correlation matrix for all numerical variables
# and find out which columns are highly correlated to each other.
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
wine_df = pd.read_csv(r'C:\Datascience datasets\winequality-red.csv')
wine_df.head()
```

```python
print(wine_df.isnull().sum())
```

```python
wine_df.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(wine_df.corr(),xticklabels=wine_df.corr().columns.values,yticklabels=wine_df.corr().columns.values,annot=True)
plt.xticks(rotation=50)
plt.show()
```

```python
# 3. Plot subplots for categorical features in titanic dataset , consisting of rows = 3 and columns = 2 format.

tit_df=pd.read_csv(r'C:\Datascience datasets\titanic.csv')
tit_df.head()
```
